## matplotlib.pyplot

- 데이터 분석의 꽃, 시각화 방법에서 가장 많이 사용되는 라이브러리인 matplotlib을 배워봅니다.


- matplotlib은 파이썬에 다양한 그래프 작성방법을 제공합니다.


- seaborn, folium 같은 다른 다양한 시각화 라이브러리들은 matplotlib을 dependency로 가지고 있거나, 영향을 많이 받았습니다.


- matplotlib은 figure()를 기본적으로 그래프 그리는 객체로 사용하고, plot(), scatter() 함수를 이용하여 원하는 그래프를 그립니다.

```python
# matplotlib은 이렇게 불러오는 것이 관행입니다.
import matplotlib.pyplot as plt
import seaborn as sns

# 이제 안써도 됨
```

```python
# 도화지를 깔기
plt.figure()
# 해당 리스트를 기준으로 하는 선을 그립니다.
plt.plot([1,2,3,4,5,6,7,8,9],[1,2,3,4,6,8,12,15,17])
# 보여주기
plt.show()
```

```python
# sample data를 하나 만들어봅시다!
x=['Americano','cafe_latte','coldbreu','flatwhite']
y=[1400,3200,3600,4030]
```

**도화지에 그림을 그려봅시다**

순서대로 하나씩 세팅해봅니다.

1. figure & figsize
2. barplot
3. grid
4. title
5. axis label
6. lineplot
7. set label
8. legend
9. set limit in axis

```python
# 막대 그래프 그리기
plt.figure(figsize=(8,6))
#plt.grid(alpha=0.5, color='black', linewidth=0.5)
plt.title("Coffee Sales in March", fontsize=14, loc="left")
plt.xlabel("Menu", fontsize=12,)
plt.ylabel('Sales', fontsize=12)
plt.bar(x,y, color='lightcoral', label='sales')

plt.plot([2000,3000,4000,4200], label='Google Trend')
plt.ylim(0,5000)
plt.legend(loc='best')
plt.show()
```

**이번에는 scatterplot을 그려봅시다**

1. figure & figsize
2. scatterplot
3. set color
4. set marker

```python
# 이번에는 scatterplot을 그려봅니다.
import numpy as np

x= np.random.random(100)
y=np.random.random(100)

plt.figure(figsize=(8,8))
plt.scatter(x,y, color='red', marker='>', s=50, alpha=0.5)
plt.show()
```

```python
# subplot을 이용하여 여러 개의 plot을 한번에 그립니다.
# subplot()을 이용하여 여러개의 plot을 하나의 figure에 출력할 수 있다.
plt.figure()
plt.subplot()

plt.show()

# plt.subplot(nrows, ncols, index)
# nrows와 ncols가 있다고 했을 때의 위치!
## 처음할 때 굉장히 헷갈립니다!! 2 x 2로 연습을 많이 해보세요 :)
```
